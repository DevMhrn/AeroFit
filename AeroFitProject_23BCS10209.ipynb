#**About the Case Study**

**Aerofit** is a leading brand in the field of **fitness equipment**. Aerofit provides a product range
including machines such as **treadmills**, **exercise bikes**, **gym equipment**, and **fitness
accessories** to cater to the needs of all categories of people.



**Problem Statement**

The market research team at AeroFit wants to identify the characteristics of the target audience for each type of treadmill offered by the company, to provide a better
recommendation of the treadmills to the new customers.

****
**Objective**

- Perform descriptive analytics to create a customer profile for each AeroFit treadmill product by developing appropriate tables and charts.
- For each AeroFit treadmill product, construct two-way contingency tables and compute all conditional and marginal probabilities along with their insights/impact on the business.

# Statistical Analysis

```python

gender_counter  = df['Gender'].value_counts()
total_count = gender_counter.sum()
male_count = (gender_counter['Male']/ total_count)*100
female_count = (gender_counter['Female']/ total_count)*100

sns.countplot(data=df, x = 'Gender', )

plt.text(0, gender_counter['Male'], f"{male_count:.2f}%", ha='center', va='bottom')
plt.text(1, gender_counter['Female'], f"{female_count:.2f}%", ha='center', va='bottom')
plt.title('Gender Distribution')
```



- **57.78%** of customer of the Product  are **Male**

- **42.22%** of customer of the Product are **Female**





```python

status_counter  = df['MaritalStatus'].value_counts()

##To calculate the Average

total_count = status_counter.sum()
single_count = (status_counter['Single']/ total_count)*100
partnered_count = (status_counter['Partnered']/ total_count)*100

sns.countplot(data=df, x = 'MaritalStatus' , hue='MaritalStatus', palette ={'Single':'skyblue', 'Partnered':'lime'})

plt.text(0, status_counter['Single'], f"{single_count:.2f}%", ha='center', va='bottom')
plt.text(1, status_counter['Partnered'], f"{partnered_count:.2f}%", ha='center', va='bottom')

```



- **40.56%** of customer of Treadmill Product are **Single**
- **59.44%** of customer of Treadmill Product are **Partnered**

```python

plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.title('Male')

male = df[df['Gender'] == 'Male']
gender_count = male['MaritalStatus'].value_counts()
total_count = gender_count.sum()

single_male = (gender_count['Single'] / total_count) * 100
partnered_male = (gender_count['Partnered'] / total_count) * 100

sns.countplot(data=male, x='MaritalStatus',hue='MaritalStatus', palette={'Single':'skyblue', 'Partnered':'lime'})

plt.text(0, gender_count['Single'], f"{single_male:.2f}%", ha='center', va='bottom')
plt.text(1, gender_count['Partnered'], f"{partnered_male:.2f}%", ha='center', va='bottom')


plt.subplot(1, 2, 2)
plt.title('Female')

female = df[df['Gender'] == 'Female']
gender_count = female['MaritalStatus'].value_counts()
total_count = gender_count.sum()

single_female = (gender_count['Single'] / total_count) * 100
partnered_female = (gender_count['Partnered'] / total_count) * 100

sns.countplot(data=female, x='MaritalStatus',hue='MaritalStatus', palette={'Single':'skyblue', 'Partnered':'lime'})


plt.text(0, gender_count['Partnered'], f"{partnered_female:.2f}%", ha='center', va='bottom')
plt.text(1, gender_count['Single'], f"{single_female:.2f}%", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
